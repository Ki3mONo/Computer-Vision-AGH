```python
import numpy as np
import cv2 as cv
import os 
IMAGES_DIR = "./images"
SEG_PARAMS = dict(
    b_minus_r_thresh = 30,
    b_minus_g_thresh = 20,
    b_min_val        = 80,
    dilate_k         = 4,
    dilate_iter      = 1,
)
def segment_calipers(img_bgr, p):
    b = img_bgr[:,:,0].astype(np.int16)
    g = img_bgr[:,:,1].astype(np.int16)
    r = img_bgr[:,:,2].astype(np.int16)
    mask = (
        (b - r > p['b_minus_r_thresh']) &
        (b - g > p['b_minus_g_thresh'])
        # (img_bgr[:,:,0] > p['b_min_val'])
    ).astype(np.uint8) * 255
    cv.imshow('mask', mask)
    cv.waitKey(0)
    cv.destroyAllWindows()
    mask = mask &  (img_bgr[:,:,0] > p['b_min_val'])
    kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (p['dilate_k'], p['dilate_k']))
    return cv.dilate(mask, kernel, iterations=p['dilate_iter'])


for image in os.listdir(IMAGES_DIR):
    img = cv.imread(IMAGES_DIR + "/" + image)
    mask = segment_calipers(img, SEG_PARAMS)
    dst = cv.inpaint(img,mask,3,cv.INPAINT_TELEA)
    cv.imshow('dst', dst)
    cv.waitKey(0)
    cv.destroyAllWindows()
```
