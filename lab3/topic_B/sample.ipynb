```python
import numpy as np
import cv2 as cv
import os 
IMAGES_DIR = "./images"
SEG_PARAMS = dict(
    b_minus_r_thresh = 30,
    b_minus_g_thresh = 20,
    b_min_val        = 80,
    dilate_k         = 4,
    dilate_iter      = 1,
)
def segment_calipers(img_hsv, p):
    
    h = img_hsv[:,:,0] # .astype(np.int16)
    s = img_hsv[:,:,1] # .astype(np.int16)
    v = img_hsv[:,:,2] # .astype(np.int16)
    cv.imshow('dst', h)
    cv.waitKey(0)
    cv.destroyAllWindows()
    cv.imshow('dst', s)
    cv.waitKey(0)
    cv.destroyAllWindows()
    cv.imshow('dst', v)
    cv.waitKey(0)
    cv.destroyAllWindows()
    mask = (
        (h - s > p['b_minus_r_thresh']) &
        (s - v > p['b_minus_g_thresh']) &
        (img_hsv[:,:,0] > p['b_min_val'])
    ).astype(np.uint8) * 255
    kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (p['dilate_k'], p['dilate_k']))
    return cv.dilate(mask, kernel, iterations=p['dilate_iter'])


for image in os.listdir(IMAGES_DIR):
    img = cv.imread(IMAGES_DIR + "/" + image)
    mask = segment_calipers(img, SEG_PARAMS)
    dst = cv.inpaint(img,mask,3,cv.INPAINT_TELEA)
    cv.imshow('dst', dst)
    cv.waitKey(0)
    cv.destroyAllWindows()
```
